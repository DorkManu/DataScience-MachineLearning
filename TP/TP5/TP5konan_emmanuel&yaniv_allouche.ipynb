```python
## Computation of the sum of the values of the climatic variables on M periods (M<<L, the total number of days)
dict_clim_var_train={}
dict_clim_var_test={}

#for var in ['ET', 'P', 'RG', 'TMIN', 'TMAX']:
for var in list_variables:
    dict_clim_var_train[var]=df_train[[key for key in df_train.keys() if var in key]].values    
    dict_clim_var_test[var]=df_test[[key for key in df_train.keys() if var in key]].values
len(dict_clim_var_train['ET'][0])
```

```python
# A routine to compute the sum of a daily variable x in a M periods of equal length
def get_var_period(M,x,L):
    dt=L/M
    newx=np.zeros(M)
    for j in range(L):
        if int(j/dt)==int((j+1)/dt) or j==L-1:
            newx[int(j/dt)]+=x[j]
        else:
            newx[int(j/dt)]+=((int(j/dt)+1)*dt-j)*x[j]
            newx[int(j/dt)+1]+=(j+1-(int(j/dt)+1)*dt)*x[j]
    return newx
    
```

```python
# Get the new variables of reduced dimension for $M$ periods, for both train and test sets
M=10
X_red_train=np.zeros([n_train,M*len(list_variables)])
X_red_test=np.zeros([n_test,M*len(list_variables)])
for i in range(n_train):
    for j in range(len(list_variables)):
        L=len(dict_clim_var_train[list_variables[j]][i])
        x_red_i=get_var_period(M,dict_clim_var_train[list_variables[j]][i],L)
        X_red_train[i,j*M:(j+1)*M]=x_red_i
for i in range(n_test):
    for j in range(len(list_variables)):
        L=len(dict_clim_var_test[list_variables[j]][i])
        x_red_i=get_var_period(M,dict_clim_var_test[list_variables[j]][i],L)
        X_red_test[i,j*M:(j+1)*M]=x_red_i
    
```

```python
# TODO
#linear regression based on X_train_2 and computation of the test error

# build and fit the model
linear_model = LinearRegression().fit(X_red_train, y_train)

# predict
y_train_est = linear_model.predict(X_red_train)
y_test_est = linear_model.predict(X_red_test)


# compute train and test errors
mse_train = mean_squared_error(y_train,y_train_est)
r2_train = r2_score(y_train,y_train_est)

mse_test = mean_squared_error(y_test,y_test_est)
r2_test = r2_score(y_test,y_test_est)
################################################

# printing values
print('Coeficients :' ,linear_model.coef_)                          #coefficients of linear regression based on the tuple (X_train_n, y_train)
print()
print('MSE_train {}, MSE_test: {}'.format(mse_train, mse_test))     #Error generated by estimation given by the linear regression
print()
print('R2_score_train {}, R2_score_test : {}'.format(r2_train, r2_test))  #The score of the linear equation

#Affichage graphique
plt.plot(y_train, y_train_est,"og", label = "training data")
plt.plot(y_test, y_test_est, "or", label = "testing data")
plt.legend()
plt.title('Linear regression')
plt.xlabel("true values")
plt.ylabel("values estimated")
plt.grid(True)
plt.show()
print(X_train.sum())
```

#### Feature selection with Lasso

The Lasso model needs an hyperparameter `alpha` to be set. In order to select the best value of `alpha`, we perform a cross validation on the training data, by using the `GridSearchCV` object from sklearn.

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
params = {'alpha': np.logspace(0, 2, 20)} #How can we choose these coefficients ?

# TODO
# set the grid search and train the grid of predictors
gscv = GridSearchCV(estimator=Lasso(), 
                   param_grid=params, 
                   scoring='neg_mean_squared_error')

lassocv = gscv.fit(X_train_n, y_train)
#print("MSE: ",lassocv.best_params_)

lasso_reg = Lasso(lassocv.best_params_['alpha'])
lasso_reg.fit(X_train_n, y_train)

# plot the coefficients of the best estimator. How many features were selected ?
print("best parameter : ", gscv.best_params_)
print("nombre de paramètres sélectionnés (non nuls) : ", sum ([1 for val in lasso_reg.coef_ if abs(val)!=0]))


#Affichage graphique
y_train_est = lasso_reg.predict(X_train_n)
y_test_est = lasso_reg.predict(X_test_n)

mse_test = mean_squared_error(y_test,y_test_est)
r2_test = r2_score(y_test,y_test_est)
print('MSE_train {}, MSE_test: {}'.format(mse_train, mse_test))

plt.plot(y_train, y_train_est,"og", label = "training data")
plt.plot(y_test, y_test_est, "or", label = "testing data")
plt.legend()
plt.title('Lasso regression')
plt.xlabel("true values")
plt.ylabel("values estimated")
plt.grid(True)
plt.show()  


#On constate que la regression Lasso donne de meilleures prédictions. Les prédictions sont plus regroupées autour de la droite d'équation y=x.
```
